```python
def final_text(text,df):
    #print(df)
    numbers=[] 
    for x in range(len(text)-1):
        if (text[x]=="<") & (text[x+1]=="$") :
            numbers.append(x+1)
        elif (text[x]=="$") & (text[x+1]==">"):
            numbers.append(x)
            cols=[]
    for x in range(0,len(numbers),2):
        y=x+1
        #print(text[numbers[x]+1:numbers[y]])
        cols.append(text[numbers[x]+1:numbers[y]])
    #print(df[cols])
    for x in cols:
        text=text.replace(x,str(df[x]))
    text=text.replace("<$","")
    text=text.replace("$>","")
    return text
    
```

```python
def meta_data(meta,df):
    meta=eval(meta)
    print(meta)
    meta_json={}
    for x in meta: 
        meta_json[x]=df[x]
    return meta_json
```
